# Relevo

O objetivo desde ***notebook*** é obter dados altimétricos de uma região a partir de um arquivo de texto e apresentar o gráfico de superfície deste relevo.

## Bibliotecas Utilizadas

Para a manipulação de gráficos utilizando a linguagem de programação Python, serão utilizadas as bibliotecas `NumPy` e `Matplotlib`.

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

## Os Dados

Os dados utilizados foram obtidos a partir do site do [Instituto Brasileiro de Geografia e Estatística (IBGE)](https://www.ibge.gov.br/), acessando a guia **Geociências** e escolhendo a opção **MODELOS DIGITAIS DE SUPERFÍCIE > Modelos digitais de superfície**.

```python
X, Y, Z = list(), list(), list()

with open('MDE_26124se_v1.XYZ') as file:
    for row in file:
        x, y, z = row.split(',')
        X.append(x); Y.append(y); Z.append(z)
```

## A Superfície



```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```
